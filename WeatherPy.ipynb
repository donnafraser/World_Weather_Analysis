```python
# Import dependencies

import pandas as pd

import matplotlib.pyplot as plt

import numpy as np
```

```python
# creating arrays of latitudes and longitudes, we'll declare each array as a variable
# in here, we will also use the zip() function to pack the lats and lngs as pairs by zipping them

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)

lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)

lat_lngs = zip(lats, lngs)

lat_lngs

# When we run this cell, the output is a zip object in memory
```

```python
# The following is the zip packing and unpacking tests from 6.1.4 (after adding code to API_pratice)

# Add the latitudes and longitudes to a list

coordinates = list(lat_lngs)

#You can only unzip a zipped tuple once before it is removed from the computer's memory.
# Make sure you unzip the latitudes and longitudes into the coordinates list before moving on.
```

```python
# Use the print() function to display the latitude and longitude combinations.

for coordinate in coordinates:
    
    print(coordinate[0], coordinate[1])
    
# When we run this cell, the output is ordered pairs of our lats and lngs coordinates.    
```
